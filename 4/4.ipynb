```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from scipy.fft import fft
from scipy.io import wavfile

%matplotlib inline
```

# 1


```python
n = np.arange(0, 100)
fs = 200
ts = 1/fs

x1 = np.cos(2*np.pi*20*n*ts + np.pi/4) + \
    3 * np.cos(2*np.pi*40*n*ts + 2*np.pi/5) + \
    2 * np.cos(2*np.pi*60*n*ts + np.pi/8)

fft_x1 = fft(x1, fs)
modulo_fft_x1 = np.abs(fft_x1)


f, ax = plt.subplots(figsize=(7, 5))


ax.plot(modulo_fft_x1)
ax.set(xticks=(range(0, fs+20, 20)))

plt.show()
```

# 2

```python
n = np.arange(0, 1000)
fs = 8e3

x2 = 2*np.sin(2000*np.pi * n/8000)

X = fft(x2)
N = np.abs(X).size
f = n/N * fs

fig, ax = plt.subplots(nrows=2, figsize=(7,7))

ax[0].plot(f, np.abs(X))
ax[1].plot(f, np.abs(X)**2)

plt.show()
```

# 3

```python
fs, data = wavfile.read('./romansaSmall.wav')

x3 = data[:, 1]

X = fft(x3)
N = np.abs(X).size
f = np.arange(0, N)/N * fs

plt.plot(f, np.abs(X))
plt.xlim([0, 7e3])

plt.ticklabel_format(axis='x', scilimits=(3,3), useMathText=True)
```
