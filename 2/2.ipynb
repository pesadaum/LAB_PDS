```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

%matplotlib inline
```

# 1

```python
def impseq(n0, n1, n2, plot=True):
    n = np.arange(n1, n2 + 1, dtype='float64')
    x = n - n0 == 0

    if plot:
        f, ax = plt.subplots()
        ax.stem(n, x, linefmt=':')
        return (n, x, (f, ax))

    return (n, x)
```

```python
n, x, plot = impseq(0, 0, 10)

n, x 
```

```python
n, x, plot = impseq(2, 0, 10)
```
