```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
A = np.array(
    [[1, 2, 3],
     [4, 5, 6],
     [7, 8, 0]])

det_A = np.linalg.det(A)

if det_A != 0:
    A_inv = np.linalg.inv(A)
else:
    A_inv = 'matriz singular'

A_inv
```

```python
z = 3+4j

modulo = np.abs(z)
angulo = np.angle(z)
log = np.log(z)
parte_imag = np.imag(z)
parte_real = np.real(z)
arct = np.arctan(z)


print(f'{modulo=}',
      f'{angulo=}',
      f'{log=}',
      f'{parte_real=}',
      f'{parte_imag=}',
      f'{arct=}',
      sep='\n')
```

```python
plt.plot(np.real(z), np.imag(z), 'o')
plt.xticks([3])
plt.yticks([4])
plt.xlabel(r'Parte real [$\mathcal{Re}$]')
plt.ylabel(r'Parte imaginária [$\mathcal{Im}$]')

plt.show()
```

# 4

```python
a = 2+3j
b = -2+1j
c = -2-3j
d = 1-3j

l = [a, b, c, d]

for num in l:
    plt.plot(np.real(num), np.imag(num), 'o', label=num)

plt.xticks(range(-2, 3))
plt.legend()
plt.xlabel(r'Parte real [$\mathcal{Re}$]')
plt.ylabel(r'Parte imaginária [$\mathcal{Im}$]')

plt.show()
```

```python
delta = 0.01

x1 = np.arange(0, 2 + delta, delta)
x2 = np.arange(0, 4 + delta, delta)

X1, X2 = np.meshgrid(x1, x2)

h2 = 100*(X2 - X1**2)**2 + (1 - X1)**2

plt.contourf(X1, X2, h2)  # ou plt.contour(X1, X2, h2)
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.colorbar()

plt.show()
```

# 7

```python
def pqs(x, y):
    p = 3 * np.sin(x**2) + 2 * np.cos(y**2)
    q = 3 * np.cos(x*y) + 2*y**2
    s = p + q

    return (p, q, s)


x = np.linspace(0, 5, 1000)
y = 0.05*x + 2.01

p, q, s = pqs(x, y)

plt.plot(x, p, label='p')
plt.plot(x, q, label='q')
plt.plot(x, s, label='p+q')

plt.legend()

plt.show()
```
